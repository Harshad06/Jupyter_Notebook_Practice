#### Matplotlib Tutorial 2 - format strings in plot function

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

#### Matplotlib Tutorial 3 - Axes labels, Legend, Grid

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
days = [1,2,3,4,5,6,7]     
max_t = [30,51,52,48,47,49,46]
min_t = [43,42,40,44,33,35,37]
avg_t = [45,48,48,46,40,42,41]
```

```python
plt.xlabel('Days')
plt.ylabel('Temperature')
plt.title('Weather')

plt.plot(days,max_t, label='MAX')
plt.plot(days,min_t, label='MIN')
plt.plot(days,avg_t, label='AvG')

# plt.legend()
# plt.legend(loc= 'upper right')
plt.legend(loc="best", shadow=True, fontsize="small")

plt.grid()
```

#### Matplotlib Tutorial 4 - Bar Chart

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
company = ['GOOGLE', 'AMZN', 'MSFT', 'FB']
revenue = [90, 136, 89, 27]
profit = [40,5,34,12]
```

```python
xpos = np.arange(len(company))
xpos
```

```python
plt.bar(xpos,revenue)
```

```python
plt.xticks(xpos,company)   #changes x numeric labels with corresponding company labels
plt.ylabel('Revenue (in Billion)')
plt.title("US Tech Stocks")
plt.bar(xpos-0.2,revenue, width=0.4, label='Revenue')
plt.bar(xpos+0.2,profit, width=0.4, label="Profit")

plt.legend()
```

```python
# Horizontal bar

plt.yticks(xpos,company)   #changes y numeric labels with corresponding company labels

plt.title("US Tech Stocks")
plt.barh(xpos-0.2,revenue, label='Revenue')
plt.barh(xpos+0.2,profit, label="Profit")

plt.legend()
```

#### Matplotlib Tutorial 6 - Pie Chart

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
exp_vals=[1400, 600,310,450,250]
exp_labels=['home rent', 'food', 'wifi', 'car','other utilities']
```

```python
plt.axis("equal")
plt.pie(exp_vals,labels=exp_labels,radius=1.5, autopct="%0.2f%%", shadow=True, explode=[0,0.2,0,0,0.2], startangle=30)
plt.show()
```

```python
# Pie Chart has Counter clockwise orientation. "start angle" is defined at 0 degree, which we can change as per our wish.
```

#### Matplotlib Tutorial 7 - Save Chart To a File Using savefig

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
exp_vals=[1400, 600,310,450,250]
exp_labels=['home rent', 'food', 'WIFI', 'car','other utilities']
```

```python
plt.axis("equal")
plt.pie(exp_vals,labels=exp_labels,radius=1.5, autopct="%0.2f%%", shadow=True, explode=[0,0.2,0,0,0.2], startangle=30)

plt.savefig("piechart.png", bbox_inches="tight", pad_inches=1, transparent=True, edgecolor='red')

## we can save file in other location also - 
#plt.savefig('C:\Users\A631055\OneDrive - Atos\Desktop\piechart.pdf')
```
