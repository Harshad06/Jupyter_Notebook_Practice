```python
from matplotlib import pylab
print(pylab.__version__)
```

```python
"""Using Numpy to generate random data"""
import numpy as np
x = np.linspace(0,10,25)
print(x)

y = x*x+2
print(y)

#print(np.array([x,y]).reshape(25,2))
```

```python
"""It Takes only one command to draw"""
pylab.plot(x,y, 'r')  # 'r' stands for red
```

```python
"""Drawing a subgraph"""
pylab.subplot(1,2,1) # The contents of the brackets represent (rows, columns, indexes)
pylab.plot(x,y, 'r--') # The third parameter here determines the color and line style
pylab.subplot(1,2,2)
pylab.plot(x,y, 'g*-')
```
